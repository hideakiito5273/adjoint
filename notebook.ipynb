```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 14

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- grad.dat ---
d = np.loadtxt("grad_x0_euler.dat", comments="#")
om, gA, gC= d[:,0], d[:,1], d[:,2]
ax1.plot(om, gA, label="analytical")
# ax1.plot(om, gB, "--", label="fdm")
ax1.plot(om, gC, ":",  label="Euler",color='Blue',linewidth=2)
d = np.loadtxt("grad_x0_heun.dat", comments="#")
om, gA, gC= d[:,0], d[:,1], d[:,2]
ax1.plot(om, gC, ":",  label="Heun",color='Red',linewidth=2)
ax1.grid()
# ax1.axhline(0, color="k", lw=0.5)
# ax1.axvline(2.0, color="r", lw=0.8, ls="--", label="ground truth")
ax1.set_xlabel("$X_0$"); ax1.set_ylabel("grad")
# ax1.set_title("勾配の比較");
ax1.legend()

# --- converge.dat ---
c = np.loadtxt("converge_euler.dat", comments="#")
h, eC = c[:,0], c[:,1]
ax2.loglog(h, eC, "o-",  label="Euler",color='Blue',linewidth=2)
c = np.loadtxt("converge_heun.dat", comments="#")
h, eC = c[:,0], c[:,1]
ax2.loglog(h, eC, "o-",  label="Heun",color='Red',linewidth=2)
# ax2.loglog(h, eB, "s--", label="err_C_disc")
ax2.loglog(h, h,    "k:",   label="O(h)")
ax2.loglog(h, h**2, "k--",  label="O(h²)")
ax2.set_xlabel("h"); ax2.set_ylabel("relative error")
# ax2.set_title("収束性")
ax2.legend()

plt.tight_layout()
# plt.savefig("result.png", dpi=150)
plt.show()
```
